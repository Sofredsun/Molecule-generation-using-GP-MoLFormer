Извлечение из базы данных ChEMBL данных об активности соединений для мишени CHEMBL5145: запрашивает измерения IC50 в наномолярных единицах из биохимических анализов, формирует датафрейм с идентификаторами молекул, их SMILES-структурами и значениями активности, затем очищает данные, а именно удаляет строки с отсутствующими значениями и приводит IC50 к числовому формату

```python
target_chembl_id = 'CHEMBL5145'

activity_api = new_client.activity
activities = activity_api.filter(
    target_chembl_id=target_chembl_id,
    standard_type='IC50',
    standard_relation__in=['=', '<'],
    standard_units='nM',
    assay_type='B'
).only(['molecule_chembl_id', 'canonical_smiles', 'standard_value'])

chembl_df = pd.DataFrame(list(activities))[['molecule_chembl_id', 'canonical_smiles', 'standard_value']]

# Удаление строк без SMILES или IC50
chembl_df = chembl_df.dropna(subset=['canonical_smiles', 'standard_value'])

# Конвертация
chembl_df['standard_value'] = pd.to_numeric(chembl_df['standard_value'], errors='coerce')
chembl_df = chembl_df.dropna(subset=['standard_value'])
```

```python
from google.colab import files

csv_name = "chembl.csv"
chembl_df.to_csv(csv_name, index=False, encoding="utf-8-sig")
print(f"Сохранено {len(chembl_df)} уникальных молекул в '{csv_name}'")

files.download(csv_name)
```
